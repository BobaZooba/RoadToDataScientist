```python
'''
Из предыдущего вывода видим, что некоторые категориальные переменные записаны неправильно как количественные,
присваиваем нашим псевдоколичественным переменным правильный тип с помощью метода .astype(), можно взять тип str,
можно тип object, помним, что при преобразовании в тип str значения NaN будут преобразованы в отдельную категорию,
а при преобразовании в тип object значения NaN так и останутся значениями NaN. выполнив преобразование, проверяем
'''
for i in ['TARGET', 'SOCSTATUS_WORK_FL', 'SOCSTATUS_PENS_FL', 'GENDER', 
          'EDUCATION', 'MARITAL_STATUS', 'GEN_INDUSTRY', 'GEN_TITLE',
          'ORG_TP_STATE', 'ORG_TP_FCAPITAL', 'JOB_DIR', 'REG_ADDRESS_PROVINCE',
          'FACT_ADDRESS_PROVINCE', 'POSTAL_ADDRESS_PROVINCE', 'TP_PROVINCE', 'REGION_NM',
          'REG_FACT_FL', 'FACT_POST_FL', 'REG_POST_FL', 'REG_FACT_POST_FL', 
          'REG_FACT_POST_TP_FL', 'FL_PRESENCE_FL', 'AUTO_RUS_FL', 'HS_PRESENCE_FL',
          'COT_PRESENCE_FL', 'GAR_PRESENCE_FL', 'LAND_PRESENCE_FL', 'DL_DOCUMENT_FL',
          'GPF_DOCUMENT_FL', 'FACT_PHONE_FL', 'REG_PHONE_FL', 'GEN_PHONE_FL']:
    if i in data.columns:
        data[i]=data[i].astype('str')
data.info()
```

```python
'''
Из предыдущего вывода также видим, что некоторые количественные переменные записаны неправильно как категориальные,
это произошло из-за того, что десятичным разделителем является запятая вместо точки, поэтому меняем запятую на точку и
присваиваем правильный тип. для краткости использум цепочку из методов .str.replace() и .astype().
выполнив преобразование, проверяем
'''
for i in ['PERSONAL_INCOME', 'CREDIT', 'FST_PAYMENT', 'LOAN_AVG_DLQ_AMT', 'LOAN_MAX_DLQ_AMT']:
    if i in data.columns:
        data[i] = data[i].str.replace(',', '.').astype('float')
data.info()
```

```python
# делаем импутацию пропущенных значений
'''
data['GEN_INDUSTRY'].fillna(data['GEN_INDUSTRY'].value_counts().index[0], inplace=True)
data['GEN_TITLE'].fillna(data['GEN_TITLE'].value_counts().index[0], inplace=True)
data['ORG_TP_STATE'].fillna(data['ORG_TP_STATE'].value_counts().index[0], inplace=True)
data['ORG_TP_FCAPITAL'].fillna(data['ORG_TP_FCAPITAL'].value_counts().index[0], inplace=True)
data['JOB_DIR'].fillna(data['JOB_DIR'].value_counts().index[0], inplace=True)
data['TP_PROVINCE'].fillna(data['TP_PROVINCE'].value_counts().index[0], inplace=True)

этот код громоздкий, можно сделать покомпактнее

for i in ['GEN_INDUSTRY', 'GEN_TITLE', 'ORG_TP_STATE', 'ORG_TP_FCAPITAL', 
          'JOB_DIR', 'TP_PROVINCE']:
    if i in data.columns:          
        data[i].fillna(data[i].value_counts().index[0], inplace=True)

а вот если надо пропуски в какой-то одной переменной заменить, то да, можно использовать код типа
data['GEN_INDUSTRY'].fillna(data['GEN_INDUSTRY'].value_counts().index[0], inplace=True)

или короче

data['Переменная'] = data['Переменная'].fillna('Самая часто встречающаяся категория')

но при преобразовании в тип str весь этот код, делающий импутацию модами, не нужен, импутация уже выполнена, давайте убедимся,
nan выделены в отдельную категорию, однако если используете тип object, используйте код выше
'''
data['JOB_DIR'].value_counts()
```

```python
'''
делаем импутацию средним, помним, что при импутации средним и любыми статистиками ДО разбиения на обучение и тест 
мы для вычисления среднего используем все наблюдения набора, а потом при разбиении некоторые наблюдения, импутированные 
с использованием информации всего набора, войдут у нас в контрольную выборку (выборку новых данных), 
а мы должны смоделировать наиболее близкую к реальности ситуацию, у нас есть обучающая выборка, никаких новых данных еще
нет, поэтому получается, что средние, которые мы получили при импутации на всем наборе, пришли к нам частично из "будущего"
(из новой, контрольной выборки, которой по факту еще нет), что неверно и приведет к переобучению
'''

data['WORK_TIME'].fillna(data['WORK_TIME'].mean(), 
                                    inplace=True)
'''
для проверки пропусков можно взять метод .info()
'''
data.info()
```

```python
data_dummies.drop(['TARGET_0', 'TARGET_1'], axis=1, inplace=True)
X = data_dummies.loc[:, 'AGE':'GEN_PHONE_FL_1']
```

```python
# Импортируем из модуля sklearn.model_selection функцию train_test_split
from sklearn.model_selection import train_test_split
# Выполняем разбиение на обучение и контроль
X_train, X_control, y_train, y_control = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train.head()
```

```python
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier(random_state=152)
from sklearn.metrics import roc_auc_score
tree.fit(X_train, y_train)
print("AUC на обучающей выборке: {:.3f}".format(roc_auc_score(y_train, tree.predict_proba(X_train)[:, 1])))
print("AUC на контрольной выборке: {:.3f}".format(roc_auc_score(y_control, tree.predict_proba(X_control)[:, 1])))
```

```python
'''
улучшим дерево, выполнив обрезку
'''
tree2 = DecisionTreeClassifier(random_state=152,max_depth=6, min_samples_leaf=22)
tree2.fit(X_train, y_train)
print("AUC на обучающей выборке: {:.3f}".format(roc_auc_score(y_train, tree2.predict_proba(X_train)[:, 1])))
print("AUC на контрольной выборке: {:.3f}".format(roc_auc_score(y_control, tree2.predict_proba(X_control)[:, 1])))
```

```python
'''
чтобы избежать предупреждения RuntimeWarning: overflow encountered in exp np.exp(prob, prob)
воспользуемся settings = np.seterr(all='ignore'), а чтобы предупреждение о переполнении не возникало,
делаем нормализацию данных
'''
settings = np.seterr(all='ignore')
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression().fit(X_train, y_train)
print("AUC на обучающей выборке: {:.3f}".format(roc_auc_score(y_train, logreg.predict_proba(X_train)[:, 1])))
print("AUC на контрольной выборке: {:.3f}".format(roc_auc_score(y_control, logreg.predict_proba(X_control)[:, 1])))
```

```python
from sklearn.ensemble import RandomForestClassifier
forest = RandomForestClassifier(n_estimators=500, random_state=152, n_jobs=-1)
forest.fit(X_train, y_train)
print("AUC на обучающей выборке: {:.3f}".format(roc_auc_score(y_train, forest.predict_proba(X_train)[:, 1])))
print("AUC на контрольной выборке: {:.3f}".format(roc_auc_score(y_control, forest.predict_proba(X_control)[:, 1])))
```
